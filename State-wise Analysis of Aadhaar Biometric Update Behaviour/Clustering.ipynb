```python
import pandas as pd
import re

def clean_state_name(x):
    if pd.isna(x):
        return x
    x = str(x).lower()
    x = re.sub(r'[^a-z\s]', '', x)   # remove symbols
    x = re.sub(r'\s+', ' ', x).strip()  # remove extra spaces
    return x
```

```python
df['state_clean'] = (
    df['state']
    .apply(clean_state_name)
    .map(state_mapping)
)
```

```python
df=df.sort_values(['state','district','pincode','date'])
df
```

```python
df['metric'] = df['bio_age_5_17'] + df['bio_age_17_']
```

```python
#df.to_csv('df.csv', index=False)

state_daily = (
    df.groupby(['state_clean', 'date'])['metric']
    .sum()
    .reset_index()
)
```

```python
state_features = state_daily.groupby('state_clean').agg(
    total_updates=('metric', 'sum'),
    avg_daily_updates=('metric', 'mean'),
    median_daily_updates=('metric', 'median'),
    std_daily_updates=('metric', 'std'),
    max_daily_updates=('metric', 'max'),
    zero_days_ratio=('metric', lambda x: (x == 0).mean())
).reset_index()
```

```python
if {'anomaly_change', 'anomaly_spatial', 'anomaly_iqr'}.issubset(df.columns):
    anomaly_features = df.groupby('state_clean').agg(
        pct_change_anomaly_rate=('anomaly_change', 'mean'),
        spatial_anomaly_rate=('anomaly_spatial', 'mean'),
        iqr_anomaly_rate=('anomaly_iqr', 'mean')
    ).reset_index()

    state_features = state_features.merge(
        anomaly_features,
        on='state_clean',
        how='left'
    )
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler

X = state_features.drop(columns=['state_clean'])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []

for k in range(2, 8):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_scaled)
    inertia.append(km.inertia_)

plt.plot(range(2, 8), inertia, marker='o')
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method for State Clustering")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
state_features['cluster'] = kmeans.fit_predict(X_scaled)
```

```python
cluster_summary = (
    state_features
    .groupby('cluster')
    .mean(numeric_only=True)
)

cluster_summary
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

state_features['pca_1'] = X_pca[:, 0]
state_features['pca_2'] = X_pca[:, 1]
```

```python
cluster_names = {
    0: "Low-Volume Stable States",
    1: "Moderate-Volume Consistent States",
    2: "Volatile / Anomaly-Prone States",
    3: "High-Volume Outlier States"
}

state_features['cluster_name'] = state_features['cluster'].map(cluster_names)

state_features[['state_clean', 'cluster', 'cluster_name']].sort_values('cluster')


import plotly.express as px

fig = px.scatter(
    state_features,
    x='pca_1',
    y='pca_2',
    color='cluster_name',     # 👈 human-readable clusters
    hover_name='state_clean',
    hover_data={
        'total_updates': True,
        'avg_daily_updates': ':.2f',
        'std_daily_updates': ':.2f',
        'zero_days_ratio': ':.2f',
    },
    title='State-wise Clustering based on Biometric Update Behaviour',
    width=900,
    height=600
)

fig.show()
```
